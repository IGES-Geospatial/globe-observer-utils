# Get Mosquito Mapper Data

```python
from go_utils import get_api_data

mhm_df = get_api_data("mosquito_habitat_mapper", end_date="2021-05-30")
```

```python
mhm_df
```

# Filter data

We will now use a qa filter method to get all records that have a larvae count greater than zero, has an identified genus, and has photos. 

```python
from go_utils.mhm import qa_filter

filtered_mhm_df = qa_filter(mhm_df,has_genus=True, min_larvae_count=1,has_photos=True)
```

```python
filtered_mhm_df
```

# Diagnostic Plots

We can then use plotting methods to quickly and visually verify that the filtering worked.

The `Larvae Count Distribution` graph shows that there is no entry with zero larvae, the `Entries with Genus Classifications vs No Classifications` shows that there are only entries with classified larvae, and the `Entries with Photos vs No Photos` shows that there are only entries with recorded photos.

```python
%matplotlib inline
from go_utils.mhm import diagnostic_plots

diagnostic_plots(filtered_mhm_df)
```

# Save as CSV

Now that we know we have the data we want, we can save it as a csv

```python
filtered_mhm_df.to_csv("MHM_Filtered.csv")
```
